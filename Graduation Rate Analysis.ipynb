```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# Importing the data
raw_data = pd.read_csv('graduation_rate.csv')
```

```python
# Renaming columns to display better
renamed_columns = []
for i in range(df.shape[1]):
    renamed_columns.append(df.columns[i].title())
renamed_columns
```

```python
df.columns = renamed_columns
df.columns
```

```python
df['Parental Level Of Education'].unique()
```

```python
# Distribution acc to parent level of education
plt.figure(figsize=(10,5))
sns.countplot(x= df['Parental Level Of Education'], data=df, 
              palette=['midnightblue','black','darkblue','red','lightblue','purple'])
plt.xticks(rotation=10, fontweight='bold')
plt.title("No of graduates according to their Parent's level of education", fontweight='bold', fontsize='large')
plt.show()
```

```python
# Distribution acc to parent's income
plt.figure(figsize=(10,5))
plt.hist(df['Parental Income'], bins=5, color='k')
plt.xticks(rotation=10, fontweight='bold')
plt.title("No. of graduates according to their Parent's Income", fontweight='bold', fontsize='large')
plt.show()
```

```python
df.head()
```

```python
# Years taken to graduate according to their Parent's level of education and income
plt.figure(figsize=(10,5))
sns.catplot(data=df,x='Years To Graduate', y='Parental Income',hue='Parental Level Of Education')
plt.xticks(rotation=90, fontweight='bold')
plt.title("Years taken to graduate according to their Parent's level of education and income", fontweight='bold', fontsize='large')
plt.show()
```

```python
# Distribution acc to high school gpa
plt.figure(figsize=(10,5))
plt.hist(df['High School Gpa'], bins=5, color='k')
plt.xticks(rotation=10, fontweight='bold')
plt.title("No. of graduates according to their High School Gpa", fontweight='bold', fontsize='large')
plt.show()
```

```python
# Distribution acc to Sat Total Score
sns.set_style('white')
plt.figure(figsize=(10,5))
plt.hist(df['Sat Total Score'], bins=10, color='green')
plt.xticks(fontweight='bold')
plt.title("No. of graduates according to their Sat Total Score", fontweight='bold', fontsize='large')
sns.despine()
plt.show()
```
